```python
import datetime, requests, yfinance
from alpaca_trade_api.rest import REST
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
API_KEY = os.getenv("ALPACA_API")
SECRET_KEY = os.getenv("ALPACA_SECRET")
```

```python
api = REST(key_id=API_KEY, secret_key=SECRET_KEY, base_url='https://paper-api.alpaca.markets')
api
```

```python
news = api.get_news('TSLA', start='2022-05-01', end='2022-05-09', include_content=True)
news
```

```python
for story in news:
    print(story.headline)
    print(story.summary)
    for image in story.images:
        print(image['url'])
```
